```python
#Importar librerías
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#Parámetros de la señal
```

```python
señal_digital=[0,1,1,0,0,1,0,1,1,1,0,1]   #Señal digital 
```

```python
número_muestras=200     #número de muestras
```

```python
#Toma de Muestras
```

```python
señal_muestreada=[]
```

```python
for i in range(1,12):    #Porque tenemos 12 dígitos en la señal digital
    u=np.ones(200)  #Vector lleno de unos que son los pulsos para tomar muestras
    m=u*señal_digital[i]    #muestras
    señal_muestreada=np.concatenate((señal_muestreada,m))
```

```python
plt.subplot(4,1,1)
plt.plot(señal_muestreada)    #Esta es la señal modulante (binaria)
```

```python
#señal sinusoidal de la portadora:
```

```python
dimension=len(señal_muestreada)   #es basicamente la longitud del vector señal_muestreada
```

```python
t=np.linspace(0,5,dimension)   #periodo de la portadora
```

```python
frec=5   #frecuencia
```

```python
portadora_s=np.cos(2*np.pi*frec*t)   #Señal portadora s
```

```python
portadora_m=np.cos(2*np.pi*frec*t + np.pi)   #Señal portadora m
```

```python
plt.subplot(4,1,2)
plt.plot(t,portadora_s)
plt.subplot(4,1,3)
plt.plot(t,portadora_m)
```

```python
#Señal modulada:
```

```python
señal_modulada =[]
for i in range(0,dimension):
    if señal_muestreada[i]==0:
        posicion_cero=np.array([portadora_s[i]])
        señal_modulada=np.concatenate((señal_modulada,posicion_cero))
    else:
        posicion_uno=np.array([portadora_m[i]])
        señal_modulada=np.concatenate((señal_modulada,posicion_uno))
```

```python
plt.subplot(4,1,4)
plt.plot(t,señal_modulada)
```

```python
plt.subplot(4,1,1)
plt.plot(señal_muestreada)  
plt.subplot(4,1,2)
plt.plot(t,portadora_s)
plt.subplot(4,1,3)
plt.plot(t,portadora_m)
plt.subplot(4,1,4)
plt.plot(t,señal_modulada)
```

<h5>1) ¿En qué consiste la modulación PSK?</h5>

R: <p>Significa "Phase Shift Keying", o Modulacion por Desplazamiento de Fase. Consiste en la modificacion de la fase de la señal. Es un metodo de modulacion de señales digitales, donde la amplitud de la señal moduladora es digital, es decir, unicamente 0 o 1.</p>
   <p>En este metodo la señal depende de dos fases portadoras con fases una a 180° grados de la otra, cada una correspondiente a un valor de la señal moduladora (0 o 1). La frecuencia y amplitud de la señal modulada se mantienen constantes.</p>
   
<h5>2) En base a todas las modulaciones vistas analógicas (Am y FM) y digitales (ASK, FSK, PSK) redacta 3 conclusiones sobre la importancia de la modulación para transmitir información.</h5>

R:

a) Los sonidos que nosotros manejamos son de baja frecuencia, pero las señales baja frecuencia tienen poco alcance, por lo que es necesario llevarlas a altas frecuencias para transmitirlas a largas distancias. La modulacion las trae de vuelta a una frecuencia legible.

b) Para enviar o recibir varias señales a la vez, es necesario enviarlas en diferentes frecuencias para que unas no interfieran con otras. Una vez mas, la modulacion las trae a todas de vuelta a una frecuencia legible.

c) Es importante modular las señales para protegerlas de ruido/interferencia. Al separar diferentes tipos de señales, tales como video o audio, en diferentes rangos, se puede evitar que todas tengan que compartir una pequeña cantiadad de frecuencias. De esta forma se puede esperar que la frecuencia de cada señal esté relativamente limpia de otras señales que pudieran interferir.
